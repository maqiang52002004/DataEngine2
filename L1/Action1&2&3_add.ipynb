Action2: 统计全班的成绩，班里有5名同学，现在需要你用Python来统计下这些人在语文、英语、数学中的平均成绩、最小成绩、最大成绩、方差、标准差。然后把这些人的总成绩排序，得出名次进行成绩输出（可以用numpy或pandas）

```python
# 1)使用Numpy
import numpy as np
persontype = np.dtype({'names':['姓名', '语文', '数学', '英语'],'formats':['S32','i', 'i', 'i']})
peoples = np.array([('Zhangfei','68','65',30),('Guangyu',95,76,98),('Liubei',98,86,88),('Dianwei',90,88,77),('Xuzhu',80,90,90)],dtype=persontype)
chineses = peoples['语文']
maths = peoples['数学']
englishs = peoples['英语']
total=peoples['语文']+peoples['数学']+peoples['英语']
print(peoples)
print('------------------------------------------------')
print(chineses,maths,englishs,total)
```

```python
print('语文，数学，英语平均成绩:',np.mean(chineses),np.mean(maths),np.mean(englishs))
print('语文，数学，英语最好成绩:',np.max(chineses),np.max(maths),np.max(englishs))
print('语文，数学，英语最差成绩:',np.min(chineses),np.min(maths),np.min(englishs))
#在numpy中计算标准差时，括号内要指定ddof的值，ddof表示自由度，当ddof=0时计算的是总体标准差；
#当ddof=1时计算的是样本标准差，当不为ddof设置值时，其默认为总体标准差。
print('语文，数学，英语成绩标准差:',np.std(chineses,ddof=1),np.std(maths,ddof=1),np.std(englishs,ddof=1))
print('语文，数学，英语成绩方差:',np.var(chineses),np.var(maths),np.var(englishs))
print('语文，数学，英语总成绩排序:',np.sort(total))
```

Action3: 对汽车质量数据进行统计
数据集：car_complain.csv
600条汽车质量投诉
Step1，数据加载
Step2，数据预处理 拆分problem类型 => 多个字段
Step3，数据统计 对数据进行探索：品牌投诉总数，车型投诉总数,哪个品牌的平均车型投诉最多


```python
import pandas as pd
#加载汽车质量数据
df =  pd. read_csv('./car_complain.csv')
df.head()
```

```python
#质量问题数据提取, 0代表没有，1代表有质量问题，Drop就是去掉这个列，get就是加上后面的列，并用逗号分隔开
#0代表没有质量问题，1代表有这个质量问题
#.get_dummies(',')按照‘，’进行分隔，是字符串的一个函数，就是onehot编码
df = df.drop('problem', axis=1).join(df.problem.str.get_dummies(','))  
def f(x):
    x=x.replace('一汽-大众','一汽大众')
    return x
df['brand']=df['brand'].apply(f) #数据清洗的一种方式
result = df.groupby(['brand'])['id'].agg(['count'])
result
```

```python
#查数据标签
tags = df.columns[7:]  
result2 = df.groupby(['brand'])[tags].agg(['sum'])
result2 = result.merge(result2, left_index=True, right_index=True, how='left')
result2
```

```python
#用‘汽车品牌’投诉总数进行排序，从大到小
df_brand = df.groupby(['brand'])['id'].agg(['count'])
df_brand= df_brand.sort_values('count' , ascending=False)
df_brand.reset_index(inplace=True)
df_brand.head(10)
```

```python
#用‘汽车品牌车型’投诉总数进行排序，从大到小
df_model = df.groupby(['car_model', 'brand'])['id'].agg(['count'])
df_model=df_model.sort_values('count' , ascending=False)
df_model.reset_index(inplace=True)
df_model
```

```python
#用‘汽车品牌车型’’投诉平均值进行排序，从大到小
import numpy as np
df_mean = df_model.groupby(['brand'])['count'].agg([np.mean])
df_mean=df_mean.sort_values('mean' , ascending=False)
df_mean.head(100)
```
